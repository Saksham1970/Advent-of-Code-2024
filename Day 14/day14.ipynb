```python
import re
pattern = r"=(?P<px>-?[0-9]*),(?P<py>-?[0-9]*).*=(?P<vx>-?[0-9]*),(?P<vy>-?[0-9]*)"
```

```python
from pydantic import BaseModel

class Robot(BaseModel):
    px: int
    py: int
    vx: int
    vy: int
```

```python
with open('day14.txt') as f:
    data = f.read()
    matches = re.finditer(pattern, data)
    robots = [Robot(**match.groupdict()) for match in matches]
```

# Part 1

```python
quads = [0, 0, 0, 0]
for robot in robots:
    px100 = robot.px + 100 * robot.vx
    py100 = robot.py + 100 * robot.vy
    
    px100 %= 101
    py100 %= 103
    
    if px100 < 50 and py100 < 51:
        quads[0] += 1
    elif px100 > 50 and py100 < 51:
        quads[1] += 1
    elif px100 < 50 and py100 > 51:
        quads[2] += 1
    elif px100 > 50 and py100 > 51:
        quads[3] += 1
    
from functools import reduce
reduce(lambda x, y: x * y, quads)
```
